# Question 1: Which is the most played chess opening?

```python
#group dataframe
popular_opening = ch.groupby('opening_name')[['id']].count()
```

```python
popular_opening = popular_opening.sort_values('id',ascending = False).reset_index() #sort values
popular_opening.rename(columns={'id':'Total','opening_name':'Opening Observed'},inplace=True) #rename columns
```

```python
#to check the first 10 rows after edits
popular_opening.head(10)
```

```python
labels=list(popular_opening['Opening Observed'])
len(labels) #calculate the length of labels
```

There are so many variations of openings played, so we will take only top 10 openings.

```python
#purely int-loc based indexing for selection by position
popular_opening = popular_opening.head(10)
popular_opening
```

```python
plt.figure(figsize=(10,5)) #width and height in inches
plt.grid() #to make the grid lines
plt.title('Popular Openings',fontsize=25,color="darkslateblue") #title of the visualization
barplot = sns.barplot(x='Total',y='Opening Observed',data=popular_opening, palette="Purples_d"); #Show point estimates as rectangular bars
barplot.set_xlabel("Total Number of Matches Played", fontsize = 20, color="darkslateblue"); #title of the x label
barplot.set_ylabel("Opening Observed", fontsize = 20,color="darkslateblue"); #title of the y label
```

# Question 2: How did most of the games concluded?

```python
#group dataframe
ending = ch.groupby('victory_status')[['id']].count()
ending.head() #show columns
```

```python
ending = ending.sort_values('id',ascending = False).reset_index() #sort values
ending.rename(columns={'id':'Total','victory_status':'Conclusion'},inplace=True) #rename columns
ending.head()  #show columns
```

```python
labels = list(ending['Conclusion']) #variable to make list
```

```python
plt.figure(figsize=(10,8)) #pie size
plt.title("Game Conclusions",fontweight='bold',fontsize=30, c='crimson') #pie title
plt.tick_params(labelsize=40,colors='r')#tickparam with labelsize and color
colors = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897'] #all colous in pie
plt.pie(ending.Total,textprops={'fontsize': 20},labels=labels,autopct='%1.0f%%',startangle=180, labeldistance=1.09,wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' }, colors=colors); #pie chart with details
#plt.legend(title = "Game Conclusions:");
```

# Question 3: Which color is likely to win more games?

```python
ch.winner.unique() #unique value
```

```python
#group dataframe
color_ch = ch.groupby('winner')[['id']].count()
color_ch
```

```python
color_ch = color_ch.sort_values('id').reset_index() #sort values
color_ch.rename(columns = {'winner':'Winner','id':'Victories'},inplace=True) #rename columns
color_ch.head()
```

```python
plt.figure(figsize=(10,5)) #plt.bar size
plt.title("Stronger side",fontweight='bold',fontsize=30,color='maroon') #title of plt.bar
plt.xlabel('Status',fontweight='bold',fontsize=30, color='maroon') #x label
plt.ylabel('Number of Victories',fontweight='bold',fontsize=25, color='maroon') #y label
plt.tick_params(labelsize=18) #labelsize
plt.grid() #for configure the grid lines
plt.bar(color_ch.Winner, color_ch.Victories,edgecolor='black',color=['darksalmon','dimgray','ivory']); #make a bar plot with x,y labels
```
